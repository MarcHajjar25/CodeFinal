```python
#https://www.bundesliga.com/en/bundesliga/stats/players
# Importing things I need in order to graph
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
# Getting the file from my computer
df = pd.read_csv('Bundisliga.csv')
df.head()
```

```python
# Getting the variables I want to use and plotting them
x = 'Goals'
y = 'Player'
plt.barh(df[y], df[x], align='center', alpha=0.5)
plt.xlabel('Goals')
plt.title('Top 5 Bundisliga Players Goals During 21-22 Season')
```

```python
# Getting the variables I want to use and plotting them
a = 'Assists'
y = 'Player'
plt.barh(df[y], df[a], align='center', alpha=0.5)
plt.xlabel('Assists')
plt.title('Top 5 Bundisliga Players Assists During 21-22 Season')
```

```python
# Getting the variables I want to use and plotting them
s = 'Shots'

plt.barh(df[y], df[s], align='center', alpha=0.5)
plt.xlabel('Shots')
plt.title('Top 5 Bundisliga Players Shots During 21-22 Season')
```

```python
# Getting the variables I want to use and plotting them
m = 'MP'
plt.barh(df[y], df[m], align='center', alpha=0.5)
plt.xlabel('Minutes Played')
plt.title('Top 5 Bundisliga Players Minutes Played During 21-22 Season')
```

```python
# Making a graph with more than one variable and then plotting it as a bar graph
import numpy as np 
import matplotlib.pyplot as plt 
  
X = list(df[y])
goals = df[x]
assists = df[a]
shots = df[s]
minutes = df[m]
  
X_axis = np.arange(len(X))
  
plt.figure(figsize = (10,10))
plt.bar(X_axis - 0.2, goals, 0.2, label = 'Goals')
plt.bar(X_axis + 0.2, assists, 0.2, label = 'Assists')
plt.bar(X_axis, shots, 0.2, label = 'shots')
  
plt.xticks(X_axis, X)
plt.xlabel("Players")
plt.ylabel("Amount")
plt.title("Top 5 Bundesliga Players Goals, Assists, and Shots During 21-22 Season")
plt.legend()
plt.show()
```
