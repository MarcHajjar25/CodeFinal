```python
#https://www.espn.com/soccer/stats/_/league/ITA.1/season/2021
# Importing modules I need in order to make graphs and stuff
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

df = pd.read_csv('SerieA.csv')
df.head()
```

```python
# Getting variables and making a graph with it
x = 'Goals'
y = 'Player'
plt.barh(df[y], df[x], align='center', alpha=0.5)
plt.xlabel('Goals')
plt.title('Top 5 Serie A Players Goals During 21-22 Season')
```

```python
# Getting variable and making a graph with it
a = 'Assists'

plt.barh(df[y], df[a], align='center', alpha=0.5)
plt.xlabel('Assists')
plt.title('Top 5 Serie A Players Assists During 21-22 Season')
```

```python
# Getting variable and making a graph with it
s = 'Shots'

plt.barh(df[y], df[s], align='center', alpha=0.5)
plt.xlabel('Shots')
plt.title('Top 5 Serie A Players Shots During 21-22 Season')
```

```python
# Getting variable and making a graph with it
m = 'MP'
plt.barh(df[y], df[m], align='center', alpha=0.5)
plt.xlabel('Minutes Played')
plt.title('Top 5 Serie A Players Minutes Played During 21-22 Season')
```

```python
# Importing modules for making a bar graph 
import numpy as np 
import matplotlib.pyplot as plt 
# List of variables I need from df
X = list(df[y])
goals = df[x]
assists = df[a]
shots = df[s]
minutes = df[m]
  
X_axis = np.arange(len(X))
  # Plotting it 
plt.figure(figsize = (10,10))
plt.bar(X_axis - 0.2, goals, 0.2, label = 'Goals')
plt.bar(X_axis + 0.2, assists, 0.2, label = 'Assists')
plt.bar(X_axis, shots, 0.2, label = 'shots')
  # Labels for graph
plt.xticks(X_axis, X)
plt.xlabel("Players")
plt.ylabel("Amount")
plt.title("Top 5 Serie A Players Goals, Assists, and Shots During 21-22 Season")
plt.legend()
plt.show()
```
