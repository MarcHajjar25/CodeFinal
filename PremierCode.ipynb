```python
#https://www.premierleague.com/stats/top/players/goals
# Importing modules I need in order to make graphs and stuff
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
# Importing file from my desktop
df = pd.read_csv('premier.csv')
df.head()
```

```python
# Making a bar graph with the variables
x = 'Goals'
y = 'Player'
plt.barh(df[y], df[x], align='center', alpha=0.5)
plt.xlabel('Goals')
plt.title('Top 5 Premier League Players Goals During 21-22 Season')
```

```python
# Getting variable and making a graph with it
a = 'Assists'
y = 'Player'
plt.barh(df[y], df[a], align='center', alpha=0.5)
plt.xlabel('Assists')
plt.title('Top 5 Premier League Players Assists During 21-22 Season')
```
